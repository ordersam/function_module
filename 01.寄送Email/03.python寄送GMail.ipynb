# python寄Gmail

```python
import json
from email.mime.text import MIMEText
import smtplib

# # 帳號密碼寫入json檔
# data = {'account': 'xxxxx@gmail.com', 'password': 'xxxxxx'}
# ret = json.dumps(data)
# with open('data/Account.json', 'w') as fp:
#     fp.write(ret)

# 讀取帳密
account = json.load(open('data/Account.json', 'r', encoding='utf-8'))
gmailUser = account['account']
gmailPasswd = account['password']

# 選擇要寄給誰(type為字串，逗號隔開)
    # to = [account['Account'], 'aaa@gmail.com', 'bbb@gmail.com']
to = [account['account']]

# 寄件內容
message = MIMEText('寶寶測試～收到請哈囉！', 'plain', 'utf-8')
message['Subject'] = 'Single Test'
message['From'] = gmailUser
message['To'] = ','.join(to)

# Set smtp：使用 Gmail 的主機
smtp = smtplib.SMTP("smtp.gmail.com:587")
    # 回應傳送伺服器
smtp.ehlo()
    # 要求加密
smtp.starttls()
    # 登入帳密
smtp.login(gmailUser, gmailPasswd)

# 寄信
smtp.sendmail(message['From'], message['To'], message.as_string())
print('Send mails OK!')
```
